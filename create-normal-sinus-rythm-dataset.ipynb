```python
def load_challenge_data(filename):
    x = loadmat(filename)
    data = np.asarray(x['val'], dtype=np.float64)
    new_file = filename.replace('.mat','.hea')
    input_header_file = os.path.join(new_file)
    with open(input_header_file,'r') as f:
        header_data=f.readlines()
    return data, header_data
```

```python
import os
ecg_filenames = []
for dirname, _, filenames in os.walk('CPSC'):
    for filename in filenames:
        filepath = os.path.join(dirname, filename)
        if filepath.endswith(".mat"):
            _, header_data = load_challenge_data(filepath)


            for i in header_data[15][5:-1].split(","):
                if i == "426783006":
                    ecg_filenames.append(filepath)
                    print(header_data)
```

```python
only_sinus = []
for i in tqdm(ecg_filenames):
    _, header_data = load_challenge_data(i)
    if len(header_data[15][5:-1].split(",")) == 1:
        only_sinus.append(i)
```

```python
os.mkdir("only_sinus")
```

```python
import shutil
new_dir = "only_sinus/"
for f in tqdm(only_sinus):
    mat_file = f.split("/")[-1]
    shutil.copy(f, new_dir+mat_file)
    f_root,_ = os.path.splitext(f)
    shutil.copy(f_root + ".hea", new_dir + f_root.split("/")[-1] +".hea")
```
